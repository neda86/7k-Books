```r
# Load necessary libraries
library(readr)

# Specify the path to your dataset
dataset_path <- "/content/book7k/books.csv"

# Load the dataset
book_dataset <- read_csv(dataset_path)

# Keep selected variables
selected_variables <- c("title", "subtitle", "authors", "published_year", "average_rating", "num_pages", "ratings_count")
book_dataset_selected <- book_dataset[selected_variables]

# Display the first few rows of the selected dataset
head(book_dataset_selected)
```

```r
library(ggplot2)
plot1 <- ggplot(book_dataset_selected, aes(x = average_rating, y = ratings_count, color = num_pages)) +
  geom_point() +
  labs(title = "average_rating vs ratings_count with num_pages as color")
plot1
```

```r
# Plot 2: log average_rating vs log ratings_count with num_pages as color
plot2 <- ggplot(book_dataset_selected, aes(x = log(average_rating), y = log(ratings_count), color = num_pages)) +
  geom_point() +
  labs(title = "log average_rating vs log ratings_count with num_pages as color")
plot2
```

```r
# Plot 3: average_rating vs ratings_count with published_year as color
plot3 <- ggplot(book_dataset_selected, aes(x = average_rating, y = ratings_count, color = as.factor(published_year))) +
  geom_point() +
  labs(title = "average_rating vs ratings_count with published_year as color")
plot3
```

```r
# Plot 4: log average_rating vs log ratings_count with published_year as color
plot4 <- ggplot(book_dataset_selected, aes(x = log(average_rating), y = log(ratings_count), color = as.factor(published_year))) +
  geom_point() +
  labs(title = "log average_rating vs log ratings_count with published_year as color")
```

```r
plot4
```

```r
# Install and load necessary libraries
library(ggplot2)
library(patchwork)

# Combine plots into a single slide
slide <- plot1 + plot2 + plot3+guides(color = "none") + plot4+guides(color = "none")

# Adjust the size of the plots
slide <- slide + theme(legend.position = "bottom") +
  theme(plot.title = element_text(size = 4)) +
  theme(legend.text = element_text(size = 4))

# Display the slide
slide
```
